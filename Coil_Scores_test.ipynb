# Obtiene Probabilidades para cada categoria

```python
n=5822
n_caravan=sum(df['CARAVAN'])
no_class = n - n_caravan

cuentas = df.groupby(['MOSTYPE']).size().reset_index(name='counts')
cross = pd.crosstab(df.CARAVAN,df['MOSTYPE'])

c_row = cuentas.shape[0]
c_col = cross.shape[1]
    
print(" Epsilon de {}".format('MOSTYPE'))

for i in range(0,c_col):
    nv=cuentas.iloc[i,0]
    nx_i=cuentas.iloc[i,1]
    
    ncx_i=cross.iloc[1,i]
    ncnx_i=cross.iloc[0,i]
    
    P_class = round(Decimal(ncx_i)/Decimal(n_caravan),6)
    
    P_noclass = round(( Decimal(nx_i) - Decimal(ncx_i)) /(Decimal(n) - Decimal(n_caravan)),6)

    
    sco_class = np.log(P_class/P_noclass)
    
    print("Value:{}     N_class:{}     P_class:{}       N_noclass:{}      P_noclass:{}   ln:{}".format(nv,ncx_i,P_class, ncnx_i, P_noclass, sco_class))
        
```

# Funci√≥n para obtener scores de training

```python
scores_df=[]

def scores(campo):
    n=5822
    n_caravan=sum(df['CARAVAN'])
    no_class = n - n_caravan

    cuentas = df.groupby([campo]).size().reset_index(name='counts')
    cross = pd.crosstab(df.CARAVAN,df[campo])

    c_row = cuentas.shape[0]
    c_col = cross.shape[1]
    
    for i in range(0,c_col):
        nv=cuentas.iloc[i,0]
        nx_i=cuentas.iloc[i,1]
    
        ncx_i=cross.iloc[1,i]
        ncnx_i=cross.iloc[0,i]
    
        P_class = round(Decimal(ncx_i)/Decimal(n_caravan),6)
    
        P_noclass = round(( Decimal(nx_i) - Decimal(ncx_i)) /(Decimal(n) - Decimal(n_caravan)),6)

        sco_class = np.log(P_class/P_noclass)
       
        scores_df.append({"campo":campo,
                     "value":nv,
                     "score":sco_class})
  
```
